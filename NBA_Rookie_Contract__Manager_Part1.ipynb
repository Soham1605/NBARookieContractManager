  Let's take a look at the correlation heatmap and make sense of some of these NBA terms and how they are related.

  Strong Positive Correlation:

*   PTS & MIN & FGM & FGA & FTM & FTA & TOV & EFF
         PTS - points scored
         MIN - minites played
         FGM - field goals made
         FGA - field goals attempted
         FTA - free throws attempted
         TOV - turnovers
         EFF - efficiency

         It can inferred that the more minutes a player plays he gets the opportunity to
         get more points, field goals, free throws, et cetera. It can also be inferred
         that the player is getting these minutes because they can score that effectively.
         It should also be noted that the longer a player is on court, and more the offensive
         load on the player, the more likely he is to commit a turnover.
      

*   3PA & 3PM
         3PA - 3 pointers attempted
         3PM - 3 pointers made
          
         The more 3 pointers a player attempts, the more he makes.

*   OREB, DREB, REB
         OREB - offensive rebounds
         DREB - defensive rebounds
         REB - rebounds
         Rebounds are of two types- you can either rebound (catch a missed bucket)
         when you are on the offensive end and save the play (this is called offensive rebound),
         or do the same on the defensive end and get the ball from the opposing team
         (this is called defensive rebound).

         A good rebounder can generally rebound well on both sides of the court.
  
  Strong Negative Correlation:

*   3PA, 3PM & OREB
         A player who is attempting 3 pointers is farther from the hoop,
         and is less likely to collect a miss, which is done by players who are closer.
*   3PA, 3PM & BLK
        BLK - blocks
         A player who is attempting 3 pointers is generally a shorter player, like
         a shooting guard, and doesn't have the most defensive responisbility.
         Blocks are made by taller players closer to the hoop, generally
         centers or power forwards.
*   3PA, 3PM & FG%
         FG% - Field goals made / field goals attempted

         A player who is shooting 3 pointers is far out in the perimeter, and
         such long shots are more difficult to make. Because of the
         difficulty of their shots, their efficiency
         (field goal percentage) naturally drops.

```python
#The point of this process is to maximise the hall of famer count by including new HOFers that may not have yet been inducted
#If the precision with how the classifier works on test data is high, we can assume that all non-HOFers classified as HOFers might
#either be inducted later on, or are still All-star caliber players that our new rookies can be compared to. We do this to mitigate
#imbalance on our next dataset. Because we need more positive misclassifications, we need to keep a larger test dataset, so we will divide dataset
#into train-test ratio of 0.5

from sklearn.model_selection import train_test_split

# Split dataset into training and testing sets (50% each) while preserving class proportions
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, stratify=y, random_state=42)

print(f"Training set size: {len(y_train)}, Hall of Famers: {sum(y_train)}")
print(f"Test set size: {len(y_test)}, Hall of Famers: {sum(y_test)}")
```

```python
#We will now compare multiple models and choose the best performing on as our base classifer
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, classification_report, precision_recall_curve
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE

# Initializing classifiers, including scaling for sensitive models such as Log Reg and SVM
classifiers = {
    "Logistic Regression": Pipeline([
        ('scaler', StandardScaler()),
        ('clf', LogisticRegression(class_weight='balanced', random_state=42, max_iter=500))
    ]),
    "Random Forest": RandomForestClassifier(class_weight='balanced', random_state=42),
    "SVM (Linear Kernel)": Pipeline([
        ('scaler', StandardScaler()),
        ('clf', SVC(kernel='linear', class_weight='balanced', probability=True, random_state=42))
    ]),
    "Decision Tree": DecisionTreeClassifier(class_weight='balanced', random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "k-Nearest Neighbors": Pipeline([
        ('scaler', StandardScaler()),
        ('clf', KNeighborsClassifier())
    ]),
    "Naive Bayes": GaussianNB()
}

# Addressing imbalance with SMOTE (Optional, if class_weight alone is insufficient)
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train, y_train)

# Dictionary to store metrics
results = {}

for name, clf in classifiers.items():
    # Fitting the model
    clf.fit(X_train_balanced, y_train_balanced)

    # Predicting labels and probabilities
    y_pred = clf.predict(X_test)
    y_pred_prob = clf.predict_proba(X_test)[:, 1] if hasattr(clf, "predict_proba") else None

    # Calculating metrics
    roc_auc = roc_auc_score(y_test, y_pred_prob) if y_pred_prob is not None else None
    report = classification_report(y_test, y_pred, zero_division=0, output_dict=True)

    results[name] = {
        "Precision": report["1"]["precision"],
        "Recall": report["1"]["recall"],
        "F1-Score": report["1"]["f1-score"],
        "ROC-AUC": roc_auc
    }
```

```python
#Plotting the previous results
# Creating a DataFrame for visualization
import pandas as pd
results_df = pd.DataFrame(results).T
results_df.sort_values(by="F1-Score", ascending=False, inplace=True)

# Plotting results
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 8))
metrics = ["Precision", "Recall", "F1-Score", "ROC-AUC"]
for metric in metrics:
    plt.plot(results_df.index, results_df[metric], marker='o', label=metric)

plt.title("Classifier Performance Comparison (Scaled Data)", fontsize=16)
plt.xlabel("Classifiers", fontsize=14)
plt.ylabel("Metric Score", fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.legend()
plt.grid()
plt.show()

# Printing the sorted results
print("Classifier Performance Summary:")
print(results_df)
```

Based on the performance of all models, we will choose SVM as our base model. This is because it has a high recall and the best F1-score among all high recall models.
It also has a strong ROC-AUC value.
This is very important because we are targeting recall with a higher priority as we have very few positive rows and we need to recall those instances.

```python
#Now we are going to work on our base classifier
#We will opt in for hyperparameter tuning, boosting, stacking, bagging, threshold tuning, class weight balancing, oversampling, and voting
#We will plot the results and choose which modifications to go ahead with

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Initializing scaler
scaler = StandardScaler()

# Fiting and transforming on the training data, transforming on the test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Parameter grid for C
param_grid = {'C': [0.01, 0.1, 1, 10, 100]}

# Grid search
grid_search = GridSearchCV(SVC(kernel='linear', class_weight='balanced', probability=True, random_state=42), param_grid, scoring='f1', cv=5, verbose=2)
grid_search.fit(X_train_scaled, y_train)

# Best parameters and model
print(f"Best parameters: {grid_search.best_params_}")
best_svm = grid_search.best_estimator_

```

```python
#Boosting implementation using AdaBoost
from sklearn.ensemble import AdaBoostClassifier

# Boosted SVM
boosted_svm = AdaBoostClassifier(estimator=best_svm, n_estimators=50, learning_rate=1.0, algorithm='SAMME', random_state=42)
boosted_svm.fit(X_train_scaled, y_train)

# Performance
y_pred_boost = boosted_svm.predict(X_test_scaled)
print(classification_report(y_test, y_pred_boost, zero_division=1))
```

```python
#Bagging implementation
from sklearn.ensemble import BaggingClassifier

# Bagging SVM
bagging_svm = BaggingClassifier(estimator=best_svm, n_estimators=10, random_state=42)
bagging_svm.fit(X_train_scaled, y_train)

# Performance
y_pred_bag = bagging_svm.predict(X_test_scaled)
print(classification_report(y_test, y_pred_bag, zero_division=1))
```

```python
#Stacking SVM with other classifiers such as Logistic regression and Random forest
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Defining base learners as SVM and Random Forest
base_learners = [
    ('svm', best_svm),
    ('random_forest', RandomForestClassifier(n_estimators=100, random_state=42))
]

# Meta-classifier is Log Reg
meta_classifier = LogisticRegression()

# Stacking
stacking_model = StackingClassifier(estimators=base_learners, final_estimator=meta_classifier, cv=5)
stacking_model.fit(X_train_scaled, y_train)

# Performance
y_pred_stack = stacking_model.predict(X_test_scaled)
print(classification_report(y_test, y_pred_stack, zero_division=1))
```

```python
#we need to tune our threshold to improve recall
# Predicting probabilities
y_pred_prob = best_svm.predict_proba(X_test_scaled)[:, 1]

# Adjust threshold
threshold = 0.3
y_pred_threshold = (y_pred_prob >= threshold).astype(int)

# Performance
print(classification_report(y_test, y_pred_threshold, zero_division=1))
```

```python
#Automatically balance classes by penalizing minority misclassification

# Balanced SVM
balanced_svm = SVC(kernel='linear', C=grid_search.best_params_['C'], class_weight='balanced', probability=True)
balanced_svm.fit(X_train_scaled, y_train)

# Performance
y_pred_balanced = balanced_svm.predict(X_test_scaled)
print(classification_report(y_test, y_pred_balanced, zero_division=1))
```

```python
#We can also oversample synthetically using SMOTE
from imblearn.over_sampling import SMOTE

# SMOTE oversampling
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_scaled, y_train)

# Training with resampled data
best_svm_smote = SVC(kernel='linear', C=grid_search.best_params_['C'], probability=True)
best_svm_smote.fit(X_train_resampled, y_train_resampled)

# Performance
y_pred_smote = best_svm_smote.predict(X_test_scaled)
print(classification_report(y_test, y_pred_smote, zero_division=1))

```

```python
#Combine multiple classifiers for a majority or soft voting approach
from sklearn.ensemble import VotingClassifier

# Voting ensemble
voting_clf = VotingClassifier(
    estimators=[
        ('svm', best_svm),
        ('log_reg', LogisticRegression()),
        ('rf', RandomForestClassifier(n_estimators=100, random_state=42))
    ],
    voting='soft'
)
voting_clf.fit(X_train_scaled, y_train)

# Performance
y_pred_voting = voting_clf.predict(X_test_scaled)
print(classification_report(y_test, y_pred_voting, zero_division=1))
```

```python
#Now that we have trained all these models, let us compare their precision, recall, etc metrics


# Function to collect metrics
def evaluate_model(model, X_test, y_test, threshold=None):
    if threshold:
        y_prob = model.predict_proba(X_test)[:, 1]
        y_pred = (y_prob >= threshold).astype(int)
    else:
        y_pred = model.predict(X_test)

    report = classification_report(y_test, y_pred, zero_division=1, output_dict=True)
    roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else y_pred)
    return {
        'Precision': report['1']['precision'],
        'Recall': report['1']['recall'],
        'F1-Score': report['1']['f1-score'],
        'ROC-AUC': roc_auc
    }

# Evaluate all models
models = {
    'GridSearch SVM': best_svm,
    'Boosted SVM': boosted_svm,
    'Bagging SVM': bagging_svm,
    'Stacking': stacking_model,
    'Balanced SVM': balanced_svm,
    'SMOTE SVM': best_svm_smote,
    'Voting Ensemble': voting_clf
}

metrics = {name: evaluate_model(model, X_test_scaled, y_test) for name, model in models.items()}
print(metrics)

# Converting to DataFrame for plotting

df_metrics = pd.DataFrame(metrics).T
df_metrics.plot(kind='bar', figsize=(12, 8))
plt.title('Model Performance Comparison')
plt.ylabel('Metric Score')
plt.xlabel('Models')
plt.legend(loc='lower right')
plt.show()
```

Boosted SVM appears to be the best option overall. It balances precision and recall better than most other models, and it has a significant improvement in ROC-AUC. While its precision isn't perfect, its recall and F1-score are more balanced compared to others.
Voting Ensemble has the highest ROC-AUC but suffers severely in recall, which suggests it is not generalizing well.
Bagging SVM and Stacking both show high precision but lack recall, which is a major concern for real-world classification tasks, especially when dealing with imbalanced classes.
SMOTE SVM does slightly improve recall, but still falls short in precision and F1-score.
